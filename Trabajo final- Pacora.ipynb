CONVERTIR TODA LA BASE DE WIDE A LONG Y A WIDE PARA QUE LOS AÑOS SEAN 1 COLUMNA, Y LA SERIE SE DIVIDA EN COLUMNAS

```python

bdtotal=pd.melt(bd2new, id_vars=['Country Name','Series Name'], value_vars=['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987',
    '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996',
    '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005',
    '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
    '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'])

bdtotal

#Cambiando nombre de la variable año a Año
Año={'variable':'Año'}
bdtotal.rename(columns=Año,inplace=True)
bdtotal

#Convertir a numericas la columna 'value'
bdtotal['value'] = pd.to_numeric(bdtotal['value'], errors='coerce')
bdtotal['value'] = bdtotal['value'].astype(float)

#base de datos total ya en wide con las variables de series y año en columnas
bdtotal2=pd.pivot_table(bdtotal, columns='Series Name', index=['Country Name','Año'], values='value').reset_index()
bdtotal2
```

```python
#Grafico de Perú -Electricidad de total de poblacion a lo largo de los años

perubd=bdtotal2[bdtotal2['Country Name']=='Peru']

alt.Chart(perubd).mark_bar().encode(
    alt.X('Año'),
    alt.Y("Access to electricity (% of population)"),
    color=alt.condition(alt.datum['Access to electricity (% of population)']>=70,
                                 alt.value('green'),
                                 alt.value('red')),
    tooltip=["Access to electricity (% of population)", "Año"]
).properties(
    title={"text": "Porcentaje de acceso a electricidad en Perú 1960-2022", "color": "Black"}).configure_axisX(labelAngle=45)
```

```python
latinbd=bdtotal2[bdtotal2['Country Name']=='Latin America & Caribbean']
americabd=bdtotal2[bdtotal2['Country Name']=='North America']
eastbd=bdtotal2[bdtotal2['Country Name']=='Middle East & North Africa']
eurobd=bdtotal2[bdtotal2['Country Name']=='European Union']
sudasia=bdtotal2[bdtotal2['Country Name']=='South Asia']
world=bdtotal2[bdtotal2['Country Name']=='World']

#América Latina

latin=alt.Chart(latinbd).mark_point().encode(
    alt.X('Año'),
    alt.Y("CO2 emissions (metric tons per capita)"),color="Country Name",
    tooltip=["CO2 emissions (metric tons per capita)", "Año"]
).interactive().properties( title={"text": "Emisiones de CO2 por región a lo largo de los años 1990-2020", "color": "Black"} )

#Norte América

america=alt.Chart(americabd).mark_point().encode(
    alt.X('Año'),
    alt.Y("CO2 emissions (metric tons per capita)"),color="Country Name",
    tooltip=["CO2 emissions (metric tons per capita)", "Año"]
).interactive()

# Oriente Medio y Africa del Norte
east=alt.Chart(eastbd).mark_point().encode(
    alt.X('Año'),
    alt.Y("CO2 emissions (metric tons per capita)"),color="Country Name",
    tooltip=["CO2 emissions (metric tons per capita)", "Año"]
).interactive()

#Europa

euro=alt.Chart(eurobd).mark_point().encode(
    alt.X('Año'),
    alt.Y("CO2 emissions (metric tons per capita)"),color="Country Name",
    tooltip=["CO2 emissions (metric tons per capita)", "Año"]
).interactive()

#Asia Sur

asiasur=alt.Chart(sudasia).mark_point().encode(
    alt.X('Año'),
    alt.Y("CO2 emissions (metric tons per capita)"),color="Country Name",
    tooltip=["CO2 emissions (metric tons per capita)", "Año"]
).interactive()

```
